```python
# importing stuff for drawing plots
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats.stats import pearsonr 
```

## Getting the data

```python
# open the dataframes with the right separators
step_data = pd.read_csv("steps.csv", sep=";")
survey_data = pd.read_csv('survey.csv')

# merge the two dataframes and show the head of the dataframe
df = survey_data.merge(step_data, on = "id")
df.head()
```

## Create the variable mean steps per participant.

(Let's inspect the variable *height*. Let's start by making a histogram.)



```python
# Slicing df based on columns. 
steps = df.loc[: , '20-6-2013':'13-5-2014']

# Add mean steps column and show the head of the dataframe
df['mean_steps'] = steps.mean(axis=1)
df.head(5)

```

## Investegate the relation of meansteps per participant with at least 2 other quantitative variables.

Let's inspect the variable height and weight.
1. Correlation matrix including steps, height and weight.
2. Two scatter plots with steps and weight and steps and height.
3. Pearsonâ€™s correlation for the 2 relations. 
4. Verbal description of the relations, including strength of the association, linearity, etc.


```python
#  Subset df and remove non plausible rows. 
df_two = df[df['weight']>45]
df_three = df_two[df_two['weight']<155]
df_four = df_three[df_three['mean_steps']<50000]
df_four.dropna()
df_five=df_four[['mean_steps', 'weight', 'height']]







```

#### Correlation matrix including steps, height and weight.

```python
sns.pairplot(df_five)
plt.show() 
```

### Scatter plot STEPS and WEIGHT & Pearsons Correlation.

```python
sns.lmplot(x="weight", y="mean_steps", data=df_five, fit_reg=False)
plt.title('Correlation between mean steps and weight')
plt.show()

print(pearsonr(df_five['weight'],df_five['mean_steps']))
```

<b> - 0.051952 is close to zero, so I can assume that there is no correlation between the average amount of steps and a persons' weight. In the scatterplot you see a cloud of datapoints, not a line. The significance of 0.1181361 indicates how likely the correlation found in the sample is, given no correlation in the population. Not likely.  <b/>
    

### Scatter plot STEPS and HEIGHT & Pearsons Correlation.

```python
sns.lmplot(x="height", y="mean_steps", data=df_five, fit_reg=False)
plt.show()

df_five[['height', 'mean_steps']].corr(method='pearson')
```

## Investegate the relation of mean steps per participant with a qualitative variable with two levels.

Let's inspect the variable city and gender.
1. Graphs of the relation (e.g. histogram, violin plot, bar plot)
2. Verbal description of the relation

```python
df_city = df_four[['mean_steps', 'city']]

sns.displot(df_city, x="mean_steps", y="city")
plt.title('Relation between City and Steps')
plt.xlabel('Steps')
plt.show()

```

```python
sns.violinplot(x="city", y="mean_steps", data=df_city)
plt.show()
```

```python
sns.displot(df_city, x="mean_steps", hue="city")
plt.show()
```

<b> Most data is collected from people living in Lille. The amount of data collected in Montpellier, Bordeaux and Lyon is almost the same. 
Most people take around 5000 steps a day on average. 
Apart from the amount of people who anticipated in the data collection there is no relation noticeable relation between the city people live in and the amount of steps they take. <b/>


```python
df_gender = df_four[['mean_steps', 'gender']]

sns.displot(df_gender, x="mean_steps", y="gender")
plt.title('Relation between Gender and Steps')
plt.xlabel('Steps')
plt.show()
```

```python
sns.violinplot(x="gender", y="mean_steps", data=df_gender)
plt.show()
```

```python
sns.displot(df_gender, x="mean_steps", hue="gender")
plt.show()
```

<b> The difference between Male and Female is small, but the graphs show that those who take a lot of steps on a day are mostly Female.  <b/>

## Investegate the relation between two qualitative variables.

Let's inspect Gender and Education.

1. Graphs of the relation.
2. Cross table with relevant percentages.
3. Verbal description of the relation

```python
# Subset df
df_education = df_four[['education_1', 'gender']]
```

```python
# create and print crosstable 
cross_tab = pd.crosstab(df_education.gender, df_education.education_1).apply(lambda r: r/r.sum()*100, axis=1)
cross_tab
```

```python
cross_tab.plot.bar(stacked=True)
plt.show()
```

<b> Both groups are well educated. 
    More males have a master or doctoratef: 51% against 36% of the females. 
    More females have a bachelor as highest degree: 42% agianst 34% of the males.<b/>
